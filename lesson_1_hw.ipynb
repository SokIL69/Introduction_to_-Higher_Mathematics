```python
# Задание 4

#!/usr/bin/env python
# -*- coding: UTF-8 -*-

import math
import numpy as np
import pylab
from matplotlib import mlab
```

```python
# Будем рисовать график функции y=cos(kx)
def func (x, k):
    """ y=cos(kx) """

    return math.cos(k * x)
```

```python
# Интервал изменения переменной по оси X
xmin = -5
xmax = 5

# Шаг между точками
dx = 100

# !!! Создаём список координат по оиси X на отрезке [-xmin; xmax], включая концы
xlist = np.linspace(xmin, xmax, dx)


# Вычислим значение функции в заданных точках
k=1
ylist1 = [func (x, k) for x in xlist]
k=5
ylist2 = [func (x, k) for x in xlist]

# !!! Рисуем графики
pylab.plot(xlist, ylist1)
pylab.plot(xlist, ylist2)

# !!! Показываем окно с нарисованными графиками
pylab.show()
```
